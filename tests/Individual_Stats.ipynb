# Individual Charts

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../Outputs/Output2017-01-16"
```

```python
df = pd.read_csv("{}.csv".format(path))
```

```python
namedf = df.set_index(keys=df.name)
```

```python
name = "REMEDIOS KENYON"
```

```python
namedf.where(namedf.home_get<100)
```

```python
namedf.ix[name].story
```

```python
name = namedf.ix[name][['inspection_put',
 'inspection_get',
 'money_search_start',
 'money_search_stop',
 'claim_put',
 'claim_get',
 'assistance_put',
 'assistance_get',
 'assessment_put',
 'assessment_get',
 'permit_put',
 'permit_get',
 'home_put',
 'home_get'
]]
```

```python
%matplotlib inline
plt.figure(figsize=(10,10))
sns.stripplot(y=name.index, x=name, jitter=True, size = 30, linewidth=1)
```
